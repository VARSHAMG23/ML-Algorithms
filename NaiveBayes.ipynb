```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart (1).csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['target'].value_counts()
```

```python
df.describe()
```

```python
# data splitting
X=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train,X_test,y_train,y_test
```

```python
# Model Building
# Guassian NaiveBayes /Normal Distribution - values are continious
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
pred1=model.predict(X_test)
```

```python
pred1
```

```python
# Metric
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_1 = accuracy_score(y_test, pred1)
print(f"Accuracy Score: {accuracy_1}")
```

```python
# 2. Multinomial NB: free , offer ,lottery (count fake words) regarding word counts / count features
from sklearn.naive_bayes import MultinomialNB
model2=MultinomialNB()
model2.fit(X_train,y_train)
```

```python
pred2=model2.predict(X_test)
```

```python
# metric
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_2 = accuracy_score(y_test, pred2)
print(f"Accuracy Score: {accuracy_2}")
```

```python
# Bernouli NB: Binary Classification  (0/1 yes/NO)---> spam
from sklearn.naive_bayes import BernoulliNB
model3=BernoulliNB()
model3.fit(X_train,y_train)
```

```python
pred3=model3.predict(X_test)
```

```python
# metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_3 = accuracy_score(y_test, pred3)
print(f"Accuracy Score: {accuracy_3}")
```

```python
# catrgorical NB : yes/no sunny/cloudy non numerical
from sklearn.naive_bayes import CategoricalNB
model4=CategoricalNB()
model4.fit(X_train,y_train)
```

```python
pred4=model4.predict(X_test)
```

```python
# Metric
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_4 = accuracy_score(y_test, pred4)
print(f"Accuracy Score: {accuracy_4}")
```
