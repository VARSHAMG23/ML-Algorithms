```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('drug.csv')
df.head(1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le  = LabelEncoder()
for i in df.columns:
  if df[i].dtype =='object':
    df[i] = le.fit_transform(df[i])
```

```python
x = df.drop('Drug', axis=1)
y = df['Drug']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVC
```

```python
model1=SVC()
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.svm import SVC
model2=SVC(kernel='linear')
model2.fit(x_train,y_train)
```

```python
y_pred1=model2.predict(x_test)
print(accuracy_score(y_test,y_pred1)*100)
# Overfitting because it is linear
```

```python
#rbf function

from sklearn.svm import SVC

model3= SVC(kernel ='rbf')

model3.fit(x_train, y_train)


y_pred2 = model3.predict(x_test)

print("accuracy score of basic svc", accuracy_score(y_pred2, y_test)*100)
```

```python
#polynomial function

from sklearn.svm import SVC

model4= SVC(kernel ='poly')

model4.fit(x_train, y_train)


y_pred3 = model4.predict(x_test)

print("accuracy score of basic svc", accuracy_score(y_pred3, y_test)*100)
```
