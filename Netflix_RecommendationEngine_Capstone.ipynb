This cell loads the Netflix dataset from a compressed CSV file into a pandas DataFrame. It specifies no header, custom column names ('Cust_Id', 'Rating'), and uses only the first two columns.

```python
df=pd.read_csv('/content/drive/MyDrive/Netflix/Copy of combined_data_1.txt.zip',header=None,names=['Cust_Id','Rating'],usecols=[0,1])
df
```

This cell calculates the number of movie entries in the dataset. These entries are identified by a `NaN` value in the 'Rating' column (as 'Cust_Id' for movies contains a colon, not a rating).

```python
# Movie count
movie_count=df.isnull().sum()['Rating']
movie_count
```

This cell determines the total number of unique customers by subtracting the identified movie entries from the total unique values in the 'Cust_Id' column.

```python
# Customer Count
total_unique=df['Cust_Id'].nunique()
Customer_count=total_unique-movie_count
Customer_count
```

This cell calculates the actual number of ratings provided by customers, excluding the movie identifier rows.

```python
# Ratings Count
ratings_count=df['Rating'].count()-movie_count
ratings_count
```

This cell counts the distribution of each star rating (1-5) across all movies.

```python
# stars count
stars_count=df['Rating'].value_counts()
stars_count
```

This cell iterates through the 'Cust_Id' column. If an entry contains a colon (':'), it's recognized as a movie ID, which is then extracted and assigned to all subsequent customer ratings until a new movie ID is encountered. This effectively associates each rating with its respective movie.

```python
movie_id=None
movie_list=[]

for customer in df['Cust_Id']:
  if ":" in customer:
    movie_id=int(customer.replace(":", ""))

  movie_list.append(movie_id)
```

This cell adds a new column 'Movie_ID' to the DataFrame `df`, populated with the `movie_list` generated in the previous step.

```python
df['Movie_ID']=movie_list
df
```

This cell removes all rows that contain any `NaN` values. This specifically removes the original movie identifier rows which had `NaN` in the 'Rating' column, leaving only the customer ratings.

```python
df.dropna(inplace=True)
```

This cell converts the 'Cust_Id' column from its current type (likely string or object) to an integer type for numerical operations.

```python
df['Cust_Id']=df['Cust_Id'].astype(int)
```

```python
movie_rat_count=df['Movie_ID'].value_counts()
movie_rat_count
```

This cell imports necessary components from the `surprise` library: `Reader` for parsing data, `Dataset` for handling rating data, `SVD` for the Singular Value Decomposition algorithm, and `cross_validate` for model evaluation.

```python
from surprise import Reader,Dataset,SVD
from surprise.model_selection import cross_validate
```

This cell initializes a `Reader` object. In `surprise`, a `Reader` is used to parse a file containing ratings. When loading from a pandas DataFrame, it often defaults to standard rating scales.

```python
reader=Reader()
```

This cell loads a subset of the preprocessed DataFrame (`df`) into a `surprise.Dataset` object. It uses the 'Movie_ID', 'Cust_Id', and 'Rating' columns and is limited to the first 100,000 rows to create a manageable dataset for model training.

```python
data=Dataset.load_from_df(df[['Movie_ID','Cust_Id','Rating']][:100000],reader)
```

This cell initializes the Singular Value Decomposition (SVD) algorithm from the `surprise` library. SVD is a powerful matrix factorization technique widely used in collaborative filtering for recommendation systems.

```python
model=SVD()
```

This cell performs 3-fold cross-validation on the initialized SVD model using the loaded dataset. It evaluates the model's performance based on the Root Mean Squared Error (RMSE) metric, providing an estimate of how well the model predicts ratings.

```python
cross_validate(model,data,measures=['RMSE'],cv=3)
```

This cell filters the main DataFrame `df` to isolate all the ratings provided by a specific customer with `Cust_Id` 1331154.

```python
#1331154
user_1331154=df[df['Cust_Id']==1331154]
user_1331154
```

This cell creates a copy of the `movie_title` DataFrame. This copy (`dummy_title`) will be used to store predicted ratings without altering the original movie title information.

```python
dummy_title=movie_title.copy()
```

This cell displays the `drop_movie_index`, which contains the Movie_IDs of movies that had a rating count below the previously calculated `bench_mark1`.

```python
drop_movie_index
```

This cell filters the `dummy_title` DataFrame, removing all movies whose `Movie_ID` is present in the `drop_movie_index`. This ensures that movies with very few ratings are excluded from further prediction steps.

```python
dummy_title=dummy_title[~dummy_title['Movie_ID'].isin(drop_movie_index)]
```

This cell iterates through the remaining movies in `dummy_title` and uses the trained SVD `model` to predict a rating for each movie for the specific user `1331154`. These predicted ratings are then stored in a new 'Rating' column within the `dummy_title` DataFrame.

```python
est_rating=[]
for movie in dummy_title['Movie_ID']:
  rating=model.predict(1331154,movie).est
  est_rating.append(rating)

dummy_title['Rating']=est_rating
```

This cell defines a function `get_user_predictions` that encapsulates the logic for generating movie recommendations for a given user. It filters out movies already rated by the user, predicts ratings for the unrated movies using the SVD model, and returns a DataFrame of movie IDs, names, and their predicted ratings. An example usage is provided for `user_id_to_predict = 1331154`.

```python
def get_user_predictions(user_id):
    # Get movies already rated by the current user
    rated_movies = df[df['Cust_Id'] == user_id]['Movie_ID'].tolist()

    # Filter out movies already rated by the user from dummy_title
    movies_to_predict = dummy_title[~dummy_title['Movie_ID'].isin(rated_movies)].copy()

    # Predict ratings for the remaining movies
    est_rating = []
    for movie_id in movies_to_predict['Movie_ID']:
        rating = model.predict(user_id, movie_id).est
        est_rating.append(rating)

    movies_to_predict['Predicted_Rating'] = est_rating
    return movies_to_predict[['Movie_ID', 'Name', 'Predicted_Rating']]


user_id_to_predict = 1331154 # You can change this to any user ID
predictions_for_user = get_user_predictions(user_id_to_predict)

print(f"Predicted ratings for user {user_id_to_predict}:")
display(predictions_for_user.head())
```

This cell demonstrates calling the `get_user_predictions` function with a different user ID (`822109`) to retrieve movie recommendations for another customer.

```python
user_id_for_new_predictions = 822109
new_predictions = get_user_predictions(user_id_for_new_predictions)

print(f"\nPredicted ratings for user {user_id_for_new_predictions}:")
display(new_predictions.head())
```
