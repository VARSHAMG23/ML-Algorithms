```python
#Import Required Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Packages related to Machine Learning
from sklearn.ensemble import RandomForestClassifier
from sklearn .model_selection import train_test_split
from sklearn.metrics import roc_auc_score,precision_score,recall_score,accuracy_score
```

```python
#Reading & Analysing data
credit =pd.read_csv('creditcard.csv')
credit.head()
```

```python
#Check for null values
credit.isnull().sum().sum()
```

```python
#check for duplicated values
credit.duplicated().sum()
```

```python
#drop duplicates if any
credit.drop_duplicates()
```

```python
credit['Target'].value_counts()
```

```python
#Machine Learning Process
X=credit.drop(columns='Target')
y=credit['Target']
```

```python
#Splitting data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=100,stratify=y)
```

```python
#creating classifier object
rf_classifier=RandomForestClassifier()
```

```python
#Train model
rf_classifier.fit(X_train,y_train)
```

```python
y_pred=rf_classifier.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
roc_auc_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix,recall_score,precision_score
cm=confusion_matrix(y_test,y_pred)
cm.ravel()
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```
